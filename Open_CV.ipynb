```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
```

```python
filename = "/content/drive/My Drive/Colab Notebooks/Image Analysis/Images/cat.jpeg"
```

```python
# the image is read in grayscale for easy processing

img = cv2.imread(filename, cv2.IMREAD_COLOR)

img
```

MATPLOTLIB uses RGB for Image visualization, while OpenCV uses BGR

```python
plt.imshow(img, cmap='gray', interpolation='bicubic')
plt.plot([50,100], [80, 100], 'c', linewidth=3)
plt.show()
```

```python
# to save an image
cv2.imwrite('cat_gray.png', img)
```

DRAWING ON IMAGES

```python
cv2.line(img, (0,0), (150,150), (255, 0, 0), 5)

cv2.rectangle(img, (10, 20), (100, 50), (255, 255, 0), 6)

cv2.circle(img, (100, 63), 55, (0, 0, 255), -1)

plt.imshow(img)
```

```python
# adding texts to images

font = cv2.FONT_HERSHEY_PLAIN

cv2.putText(img, 'OpenCV cat', (0, 130), font, 1, (200, 255, 255), 9, cv2.LINE_AA)

plt.imshow(img)
```
